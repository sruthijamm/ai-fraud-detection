```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('dark_background')
sns.set_palette('husl')

print('Imports successful')
```

```python
df = pd.read_csv('../data/creditcard.csv')
print(f'Dataset shape: {df.shape}')
print(f'\nColumn names: \n{df.columns.tolist()}')
```

```python
print('=== First 5 rows ===')
display(df.head())

print('=== Data types & nulls ===')
print(df.info())

print('\n=== Basic statistics ===')
display(df.describe())
```

```python
fraud_counts = df['Class'].value_counts()
print(fraud_counts)

print(f"\nFraud rate: {fraud_counts[1] / len(df) * 100:.4f}%")
print(f"Legit rate: {fraud_counts[0] / len(df) * 100:.4f}%")

# bar chart
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution — Fraud vs Legitimate')
plt.xticks([0, 1], ['Legitimate', 'Fraud'])
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()
```

```python
print("=== Feature breakdown ===")
print(f"PCA features: V1 through V28 (anonymized for privacy)")
print(f"Other features: Time, Amount")
print(f"Target: Class (0=legitimate, 1=fraud)")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, color='steelblue')
plt.title('Legitimate — Amount Distribution')
plt.xlabel('Amount')
plt.yscale('log')

plt.subplot(1, 2, 2)
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, color='crimson')
plt.title('Fraud — Amount Distribution')
plt.xlabel('Amount')
plt.yscale('log')

plt.tight_layout()
plt.show()
```

```python
print("=== Average Transaction Amount ===")
print(f"Legitimate: ${df[df['Class'] == 0]['Amount'].mean():.2f}")
print(f"Fraud:      ${df[df['Class'] == 1]['Amount'].mean():.2f}")

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df[df['Class'] == 0]['Time'], bins=50, color='steelblue')
plt.title('Legitimate — Time Distribution')
plt.xlabel('Seconds elapsed')

plt.subplot(1, 2, 2)
sns.histplot(df[df['Class'] == 1]['Time'], bins=50, color='crimson')
plt.title('Fraud — Time Distribution')
plt.xlabel('Seconds elapsed')

plt.tight_layout()
plt.show()
```
